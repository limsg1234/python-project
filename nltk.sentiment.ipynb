```python
import pandas as pd
import numpy as np

data = pd.read_csv('C:/one/OneDrive/Big데이터/산학프로젝트/코드/OUTPUT/output2.csv', encoding='utf-8', engine = 'python')

data = data.replace(0, np.nan, regex=True)
data['average'] = data.mean(numeric_only=True, axis=1)
data = data[['text', 'average']]

text= data['text'].values.tolist()
average = data['average'].values.tolist()


new_average = []
for i in average:
    b = 2 * i + 3
    new_average.append(b)

final = pd.DataFrame(
    {'text': text,
     'average': new_average
    })

print(final)

export_csv = final.to_csv (r'C:/one/OneDrive/Big데이터/산학프로젝트/코드/OUTPUT/output_average_scale_of_5.csv', encoding='utf-8', index = None, header=True)
```
